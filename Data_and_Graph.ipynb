```python
import pandas as pd
path="/content/healthcare-dataset-stroke-data.csv"
data=pd.read_csv(path)
print(data)
print(data.info())
print(data.isnull().sum())


median_value=data.bmi.median()
print(median_value)

data.bmi=data.bmi.fillna(median_value)
print(data)
print(data.isnull().sum())

data['gender']=data['gender'].map({'Female':0,'Male':1})
data['ever_married']=data['ever_married'].map({'No':0,'Yes':1})
data['Residence_type']=data['Residence_type'].map({'Rural':0,'Urban':1})
print(data)

median_value=data.gender.median()
print(median_value)
data.gender=data.gender.fillna(median_value)
print(data)

import sklearn
from sklearn.preprocessing import LabelEncoder
le_work_type=LabelEncoder()
le_smoking_status=LabelEncoder()

data['work_type']=le_work_type.fit_transform(data['work_type'])
data['smoking_status']=le_smoking_status.fit_transform(data['smoking_status'])

inputs=data.drop(['id','stroke'],axis=1)
print("Inputs:",inputs)
outputs=data['stroke']
print("Outputs:",outputs)
print(data.isnull().sum())



import sklearn
import math
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(inputs,outputs)

gender=str(input("enter the value"))
age=float(input("enter the value"))
hypertension=int(input("enter the value"))
heart_disease=int(input("enter the value"))
ever_married=int(input("enter the value"))
work_type=int(input("enter the value"))
Residence_type=int(input("enter the value"))
avg_glucose_level=float(input("enter the value"))
bmi=float(input("enter the value"))
smoking_status=int(input("enter the value"))

res=model.predict([[gender,age,hypertension,heart_disease,ever_married,work_type,Residence_type,avg_glucose_level,bmi,smoking_status]])
print(math.ceil(res))

acc=model.score(inputs,outputs)
print("Accuracy:",acc*100)
```
